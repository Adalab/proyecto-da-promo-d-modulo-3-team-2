> ## 1.1. Tasa de la rotación general

```python
# Calculo de las variables a graficar:
rotacion_counts = df_visualizacion["Attrition"].value_counts()
rotacion_labels = rotacion_counts.index  #'Sí' y 'No'
rotacion_sizes = rotacion_counts.values

# Estilo con Seaborn:
sns.set_style("whitegrid")
# Definición de los colores de las variables:
paletabase = ['thistle', 'peachpuff']
# Color de fondo y tamaño de la figura:
plt.figure(facecolor='snow', figsize=(9,7))

# Creación el gráfico de tarta:
plt.pie(rotacion_sizes, labels=rotacion_labels, autopct='%1.1f%%',
        colors=paletabase, 
        wedgeprops={'edgecolor': 'whitesmoke', 'linewidth': 1, 'antialiased': True},#Colores del gráfico.
        textprops={'fontsize': 14, 'bbox': dict(facecolor='whitesmoke', edgecolor='gainsboro', boxstyle='round,pad=0.2')} ,
        pctdistance=0.8, labeldistance=1.1,
        startangle=140) #Edición de etiquetas y punto de inicio del gráfico.

# Estilo del título:
plt.title("Tasa de Rotación de los Empleados", 
          fontsize=14, fontweight='normal', color='rebeccapurple',
          bbox=dict(facecolor='whitesmoke', edgecolor='gainsboro', boxstyle='round,pad=0.2'))

# Círculo central para efecto donut:
centre_circle = plt.Circle((0,0), 0.60, fc='whitesmoke', edgecolor='whitesmoke', linewidth=1.5)
plt.gca().add_artist(centre_circle)


# Ajuste del diseño y mostrar el gráfico:
plt.tight_layout()
plt.show()
```

>### 2.1.1. Edad 

* ¿Cómo influye la edad en la tasa de rotación? (Age/Attrition)

* Comparación de la distribución de edades entre empleados que se quedan y los que se van.

```python
# Creación de variable para mostrar la mediana en el segundo gráfico:
medians = df_visualizacion.groupby("Attrition")["Age"].median()

# Estilo con Seaborn:
sns.set_style("whitegrid")
# Definición de los colores de las variables:
paletabase = ['thistle', 'peachpuff']
# Creación de una figura con dos subgráficos:
fig, axes = plt.subplots(1, 2, figsize=(12, 6), facecolor="snow") #Color de fondo y tamaño de la figura.


# ----------------- 1. Gráfico de barras: Promedio de edad por grupo de dimisión ----------------- #
ax1 = axes[0] #Definición del primer gráfico.
# Creación del barplot:
barplot = sns.barplot(x=df_visualizacion["Attrition"], y=df_visualizacion["Age"], palette=paletabase, ax=ax1)

#Estilo del título y etiquetas:
ax1.set_title("Edad Promedio por Estado de Rotación", 
              fontsize=14, fontweight='normal', color='rebeccapurple',
              bbox=dict(facecolor='whitesmoke', edgecolor='gainsboro', boxstyle='round,pad=0.2'))
ax1.set_ylabel("Edad Promedio", fontsize=12, fontweight='normal')
ax1.set_xlabel("Rotación", fontsize=13, fontweight='normal')

ax1.set_ylim(0, 40)  # Ajuste del eje 'y'.
# Bucle que muestra una etiqueta sobre cada barra con su respectivo valor:
for p in barplot.patches:
    ax1.annotate(f"{p.get_height():.1f}", 
                 (p.get_x() + p.get_width() / 2., p.get_height()), 
                 ha='center', va='bottom',
                 fontsize=12,
                 bbox=dict(facecolor='whitesmoke', edgecolor='gainsboro', boxstyle='round,pad=0.2'))


# ----------------- 2. Boxplot: Distribución de edades según dimisión ----------------- #
ax2 = axes[1] #Definición del segundo gráfico.
# Creación del boxplot:
boxplot = sns.boxplot(x=df_visualizacion["Attrition"], y=df_visualizacion["Age"], palette=paletabase, ax=ax2)

# Estilo del título y etiquetas:
ax2.set_title("Distribución de Edades por Rotación", 
              fontsize=14, fontweight='normal', color='rebeccapurple',
              bbox=dict(facecolor='whitesmoke', edgecolor='gainsboro', boxstyle='round,pad=0.2'))
ax2.set_ylabel("Edad", fontsize=12, fontweight='normal')
ax2.set_xlabel("Rotación", fontsize=13, fontweight='normal')

# Bucle para mostrar la mediana:
for i, median in enumerate(medians):
    ax2.annotate(f"Mediana: {median:.1f}", 
                 (i, median), 
                 ha='center', va='bottom',
                 fontsize=12,
                 bbox=dict(facecolor='whitesmoke', edgecolor='gainsboro', boxstyle='round,pad=0.2'))

# Ajustar el diseño y mostrar gráficos
plt.tight_layout()
plt.show()
```

**Insights Clave:**
El gráfico refleja muy poca diferencia en la edad promedio de las rotaciones de personal.
Ahora bien: 
Los empleados que se quedan en la empresa (barra verde) tienen una edad promedio mayor, alrededor de 37-38 años.
Los empleados que se van de la empresa (barra roja) tienen una edad promedio menor, alrededor de 33-34 años.
Los empleados que se quedan tienen una mayor dispersión de edades, con un rango que va desde aproximadamente los 20 hasta los 60 años.

La mediana de edad en el grupo que se va es más baja que la del grupo que se queda, lo que confirma que los empleados más jóvenes abandonan con mayor frecuencia.

La rotación es mayor en empleados jóvenes. Esto puede deberse a la búsqueda de mejores oportunidades, falta de crecimiento, salarios bajos o falta de estabilidad en el empleo.
Los empleados de mayor edad tienden a quedarse. Posiblemente porque ya han encontrado estabilidad en la empresa o tienen menos incentivos para cambiar de trabajo.

**Acciones sugeridas**: Implementar estrategias de retención para empleados jóvenes, como programas de desarrollo profesional, mejores beneficios o incentivos de crecimiento.

* ¿Existen diferencias en la tasa de rotación entre hombres y mujeres? (Sex/Attrition)

```python
# Convertir las columnas relevantes a valores limpios:
df_visualizacion["Sex"] = df_visualizacion["Sex"].astype(str).str.strip()
df_visualizacion["Attrition"] = df_visualizacion["Attrition"].astype(str).str.strip()


# Estilo con Seaborn:
sns.set_style("whitegrid")
# Definición de las paletas de color:
paletabase = ['thistle', 'peachpuff']
paleta_barras = ["#FFD1D9", "#FFB58A"]

# Definición de las variables a graficar:
sex_counts = df_visualizacion["Sex"].value_counts()
attrition_by_sex = df_visualizacion.groupby("Sex")["Attrition"].value_counts(normalize=True).unstack() * 100


# Creación de la figura y los ejes con 3 subgráficos, color de fondo y tamaño:
fig, axes = plt.subplots(1, 3, figsize=(18, 6), facecolor="snow")

# ----------------- 1. Gráfico de Tarta - Distribución por Sexo ----------------- #
# Creación del gráfico de tarta:
axes[0].pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%',
            colors=paletabase,
            wedgeprops={'edgecolor': 'whitesmoke', 'linewidth': 1, 'antialiased': True},
            textprops={'fontsize': 14, 'bbox': dict(facecolor='whitesmoke', edgecolor='gainsboro', boxstyle='round,pad=0.2')},
            pctdistance=0.8, labeldistance=1.1, startangle=140) #Edición de etiquetas y punto de inicio del gráfico.

#Estilo del título y etiquetas:
axes[0].set_title("Distribución de sexo en la empresa", fontsize=14, color='rebeccapurple',
                  bbox=dict(facecolor='whitesmoke', edgecolor='gainsboro', boxstyle='round,pad=0.2'))

# Círculo central para efecto donut:
centre_circle = plt.Circle((0,0), 0.60, fc='whitesmoke', edgecolor='whitesmoke', linewidth=1.5)
axes[0].add_artist(centre_circle)

# ----------------- 2. Gráfico de Barras - Tasa de Dimisión por Sexo ----------------- #
# Creación del gráfico de barras:
attrition_by_sex[["No", "Yes"]].plot(kind="bar", color=paleta_barras, edgecolor='whitesmoke', linewidth=1, ax=axes[1])

#Estilo del título, etiquetas y leyenda:
axes[1].set_title("Tasa de Rotación por Sexo", fontsize=14, color='rebeccapurple',
                  bbox=dict(facecolor='whitesmoke', edgecolor='gainsboro', boxstyle='round,pad=0.2'))
axes[1].set_ylabel("Porcentaje (%)", fontsize=12)
axes[1].set_xlabel("Sexo", fontsize=12)
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=0)
axes[1].legend(title="Dimisión", labels=["No", "Sí"], frameon=True, facecolor='whitesmoke', edgecolor='gainsboro')

# Bucle que muestra una etiqueta sobre cada barra con su respectivo valor:
for p in axes[1].patches:
    axes[1].annotate(f"{p.get_height():.1f}%", 
                     (p.get_x() + p.get_width() / 2., p.get_height() - 2), 
                     ha='center', fontsize=12,
                     bbox=dict(facecolor='whitesmoke', edgecolor='gainsboro', boxstyle='round,pad=0.2'))

# ----------------- 3. Gráfico de Barras - Salario Promedio por Sexo ----------------- #
# Creación del gráfico de barras:
sns.barplot(x="Sex", y="Monthly_Income", data=df_visualizacion, palette=paletabase, ax=axes[2])

#Estilo del título y etiquetas:
axes[2].set_title("Salario promedio por sexo", fontsize=14, color='rebeccapurple',
                  bbox=dict(facecolor='whitesmoke', edgecolor='gainsboro', boxstyle='round,pad=0.2'))
axes[2].set_ylabel("Salario mensual promedio", fontsize=12)
axes[2].set_xlabel("Sexo", fontsize=12)

# Bucle que muestra una etiqueta sobre cada barra con su respectivo valor:
for p in axes[2].patches:
    height = p.get_height()
    axes[2].annotate(f"{height:.1f}", (p.get_x() + p.get_width() / 2., height),
                     ha='center', fontsize=12,
                     bbox=dict(facecolor='whitesmoke', edgecolor='gainsboro', boxstyle='round,pad=0.2'))

# Ajuste del diseño y mostrar los gráficos:
plt.tight_layout()
plt.show()
```

**Gráfico 3**: Salario Mensual Promedio por Sexo:

>## 2.3.  Satisfacción laboral y rotación
- ¿Cómo se relaciona la satisfacción laboral con la rotación? (JobSatisfaction/Attrition)
- Comparación de los niveles de satisfacción entre empleados que se quedan y los que se van.  

```python
# Creación de la variable que contiene la mediana:
medians = df_visualizacion.groupby("Attrition")["Job_Satisfaction"].median()

# Estilo con Seaborn:
sns.set_style("whitegrid")
# Definición de los colores de las variables:
paletabase = ['thistle', 'peachpuff']

# Crear una figura con dos subgráficos, color de fondo y tamaño:
fig, axes = plt.subplots(1, 2, figsize=(12, 6), facecolor="snow")

# ----------------- 1. Gráfico de barras: Promedio de edad por grupo de rotación ----------------- #
ax1 = axes[0] #Define el gráfico 1
# Creación del gráfico de barras:
barplot = sns.barplot(x=df_visualizacion["Attrition"], y=df_visualizacion["Job_Satisfaction"], palette=paletabase, ax=ax1)
#Estilo del título y etiquetas:
ax1.set_title("Satisfacción Laboral Promedio por Estado de Rotación", 
              fontsize=14, fontweight='normal', color='rebeccapurple',
              bbox=dict(facecolor='whitesmoke', edgecolor='gainsboro', boxstyle='round,pad=0.2'))
ax1.set_ylabel("Satisfacción Laboral Promedio", fontsize=12, fontweight='normal')
ax1.set_xlabel("Rotación", fontsize=12, fontweight='normal')
ax1.set_ylim(0.0, 4)  #Ajuste del eje 'y'.

# Bucle que muestra una etiqueta sobre cada barra con su respectivo valor:
for p in barplot.patches:
    ax1.annotate(f"{p.get_height():.1f}", 
                 (p.get_x() + p.get_width() / 2., p.get_height()), 
                 ha='center', va='bottom',
                 fontsize=12,
                 bbox=dict(facecolor='whitesmoke', edgecolor='gainsboro', boxstyle='round,pad=0.2'))

# ----------------- 2. Boxplot: Distribución de edades según rotación ----------------- #
ax2 = axes[1] #Define el gráfico 2
# Creación del boxplot:
boxplot = sns.boxplot(x=df_visualizacion["Attrition"], y=df_visualizacion["Job_Satisfaction"], palette=paletabase, ax=ax2)

#Estilo del título y etiquetas:
ax2.set_title("Distribución de Satisfacción Laboral por Rotación", 
              fontsize=14, fontweight='normal', color='rebeccapurple',
              bbox=dict(facecolor='whitesmoke', edgecolor='gainsboro', boxstyle='round,pad=0.2'))
ax2.set_ylabel("Satisfacción Laboral", fontsize=12, fontweight='normal')
ax2.set_xlabel("Rotación", fontsize=12, fontweight='normal')

# Bucle que muestra la mediana para cada variable:
for i, median in enumerate(medians):
    ax2.annotate(f"Mediana: {median:.1f}", 
                 (i, median), 
                 ha='center', va='bottom',
                 fontsize=12,
                 bbox=dict(facecolor='whitesmoke', edgecolor='gainsboro', boxstyle='round,pad=0.2'))

# Ajustar el diseño y mostrar gráficos
plt.tight_layout()
plt.show()
```

**Insights Clave:**

**Gráfico 1**: La satisfacción laboral promedio es más alta para los empleados que no se fueron ("No") en comparación con los que sí se fueron ("Yes"). Esto sugiere que la satisfacción laboral es un factor importante en la decisión de los empleados de permanecer o dejar la empresa.

**Gráfico 2**: La distribución de la satisfacción laboral es diferente entre los dos grupos. Los empleados que se fueron ("Yes") tienden a tener una distribución de satisfacción laboral más baja y más concentrada en valores bajos. Los empleados que se quedaron ("No") muestran una distribución más amplia, con una mediana más alta y algunos valores atípicos de alta satisfacción.
